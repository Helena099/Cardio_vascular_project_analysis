## V. Outlier Detection and Handling

```python
def outlier_thresholds(df, col_name, low_quantile=0.25, up_quantile=0.75):
    Q1 = df[col_name].quantile(low_quantile)
    Q3 = df[col_name].quantile(up_quantile)
    IQR = Q3 - Q1
    low_limit = Q1 - 1.5 * IQR
    up_limit = Q3 + 1.5 * IQR
    return low_limit, up_limit
```

```python
def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
```

## VI. Feature Extraction

```python
#*Interpretation*
        #Ce code crée une nouvelle colonne "Age_CAT" dans le DataFrame "Cardiovascular_Model_creation_df" en découpant la variable "Age (years)"
        #en trois catégories : "Adult" pour les individus de 20 à 40 ans, "Middle_Age_Adult" pour ceux de 40 à 60 ans, et "Senior_Adult" pour les
        #individus de plus de 60 ans. Cela permet de regrouper les données en fonction de tranches d'âge et facilite l'analyse des caractéristiques
        #ou des tendances spécifiques à chaque groupe d'âge.


Cardiovascular_Model_creation_df['Age_CAT'] = pd.cut(x=Cardiovascular_Model_creation_df['Age (years)'],
                            bins=[20, 40, 60, Cardiovascular_Model_creation_df['Age (years)'].max()],
                            labels=["Adult", "Middle_Age_Adult", "Senior_Adult"])
```
